```python
# drop rows with missing values in Ever_Married and Graduated columns
df.dropna(subset=['Ever_Married', 'Graduated'], inplace=True)

# fill missing values in Profession and Var_1 columns with mode
df['Profession'].fillna(df['Profession'].mode()[0], inplace=True)
df['Var_1'].fillna(df['Var_1'].mode()[0], inplace=True)

# fill missing values in Work_Experience and Family_Size columns with median
df['Work_Experience'].fillna(df['Work_Experience'].median(), inplace=True)
df['Family_Size'].fillna(df['Family_Size'].median(), inplace=True)
```

```python
# Convert categorical variables to numerical using LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Ever_Married'] = le.fit_transform(df['Ever_Married'])
df['Graduated'] = le.fit_transform(df['Graduated'])
df['Profession'] = le.fit_transform(df['Profession'])
df['Spending_Score'] = le.fit_transform(df['Spending_Score'])
df['Var_1'] = le.fit_transform(df['Var_1'])
df['Segmentation'] = le.fit_transform(df['Segmentation'])
```

```python
# Split the data into training and testing sets
X = df.drop('Segmentation', axis=1)
y = df['Segmentation']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Build the Decision Tree Classifier model
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = clf.predict(X_test)
```

```python
# Evaluate the model using classification report and confusion matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
